# **About the Question from the image**

```python
import base64
from openai import OpenAI

client = OpenAI(
    api_key="AIzaSyB5FWC8FVDZyQ-xnnQ19kmqzPNdFu2L3CU",
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

# Function to encode the image
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
path = r"C:\Users\abhis\OneDrive\Desktop\abhigit\SIH-Hackathon\WhatsApp Image 2025-09-16 at 22.59.36_cdd08228.jpg"
# Getting the base64 string
base64_image = encode_image(path)

response = client.chat.completions.create(
  model="gemini-2.0-flash",
  messages=[{"role" : "system", "content" : "You are a helpful assistant of a Farmer. who reply in only 100 words."},
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": "What is in this image?",
        },
        {
          "type": "image_url",
          "image_url": {
            "url":  f"data:image/jpeg;base64,{base64_image}"
          },
        },
      ],
    }
  ],
)

a = response.choices[0].message.content
cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', a)
print(cleaned_text)
```

# **Analyze audio input:**

```python
import base64
from openai import OpenAI

client = OpenAI(
    api_key="AIzaSyB5FWC8FVDZyQ-xnnQ19kmqzPNdFu2L3CU",
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
intput = "give me answer about the question in audio"
audio_file = r"C:\Users\abhis\OneDrive\Desktop\abhigit\SIH-Hackathon\WhatsApp Audio 2025-09-16 at 23.21.19_c34ce57.aac"
with open(audio_file, "rb") as audio_file:
  base64_audio = base64.b64encode(audio_file.read()).decode('utf-8')

response = client.chat.completions.create(
    model="gemini-2.0-flash",
    messages=[{"role" : "system", "content" : "You are a helpful assistant of a Farmer. who reply in only 100 words."},
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": intput,
        },
        {
              "type": "input_audio",
              "input_audio": {
                "data": base64_audio,
                "format": "wav"
          }
        }
      ],
    }
  ],
)

a = response.choices[0].message.content
cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', a)
print(cleaned_text)
```

**Use the google serach engine to find the content of farming**

```python
from google import genai
from google.genai import types
```

```python

# Configure the client
client = genai.Client(api_key="AIzaSyB5FWC8FVDZyQ-xnnQ19kmqzPNdFu2L3CU")

# Define the grounding tool
grounding_tool = types.Tool(
    google_search=types.GoogleSearch()
)

# Configure generation settings
config = types.GenerateContentConfig(
    tools=[grounding_tool]
)
ip = input()
# Defining the role based content
c = [
    types.Content(role="user", parts=[{"text": f" reply me like a farmer's asistant {ip}. Reply in 100 words"}])
]


# Make the request
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=c,
    config=config,
)

# Print the grounded response
print(response.text)
```
