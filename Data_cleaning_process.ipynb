# **Using the FT - Transformer**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt

file_path = "crop_final.csv"
df = pd.read_csv(file_path)
df
```

```python
target_col = "Crop Type"


X = df.drop(columns=[target_col])
y = df[target_col].values


cat_cols = X.select_dtypes(include=["object", "bool"]).columns.tolist()
num_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
```

```python
cat_cols
```

```python
num_cols
```

```python
# Encode categorical features
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    label_encoders[col] = le
```

```python
# Normalize numerical features
scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])
```

```python
# Convert to numpy
X = X.values
X
```

```python
y_le = LabelEncoder()
y_encoded = y_le.fit_transform(y)
```

```python
X_train, X_val, y_train_encoded, y_val_encoded = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

**DataLoader**

```python
class TabularDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32).unsqueeze(1)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = TabularDataset(X_train, y_train)
val_dataset = TabularDataset(X_val, y_val)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
```

**FT Transformer**

```python
class FTTransformer(nn.Module):
    def __init__(self, num_features, d_model=64, num_heads=4, num_layers=3, dropout=0.1):
        super().__init__()
        
        self.input_layer = nn.Linear(num_features, d_model)
        
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=num_heads,
            dropout=dropout,
            activation='gelu'
        )
        
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.fc_out = nn.Linear(d_model, 1)  # Regression

    def forward(self, x):
        x = self.input_layer(x)
        x = self.transformer(x.unsqueeze(1)).squeeze(1)
        return self.fc_out(x)
```

```python
num_features = X.shape[1]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = FTTransformer(num_features).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
train_losses, val_losses = [], []
train_r2, val_r2 = [], []

EPOCHS = 400
for epoch in range(EPOCHS):
    # ---- Training ----
    model.train()
    epoch_train_loss = 0
    y_true_train, y_pred_train = [], []

    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)

        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()

        epoch_train_loss += loss.item()
        y_true_train.extend(y_batch.cpu().detach().numpy())
        y_pred_train.extend(outputs.cpu().detach().numpy())

    # Train metrics
    epoch_train_loss /= len(train_loader)
    train_r2_score = r2_score(y_true_train, y_pred_train)

    # ---- Validation ----
    model.eval()
    epoch_val_loss = 0
    y_true_val, y_pred_val = [], []

    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            epoch_val_loss += loss.item()

            y_true_val.extend(y_batch.cpu().numpy())
            y_pred_val.extend(outputs.cpu().numpy())

    epoch_val_loss /= len(val_loader)
    val_r2_score = r2_score(y_true_val, y_pred_val)

    train_losses.append(epoch_train_loss)
    val_losses.append(epoch_val_loss)
    train_r2.append(train_r2_score)
    val_r2.append(val_r2_score)

    print(f"Epoch {epoch+1}/{EPOCHS}, Train Loss: {epoch_train_loss:.4f}, Val Loss: {epoch_val_loss:.4f}, Train R²: {train_r2_score:.4f}, Val R²: {val_r2_score:.4f}")
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Val Loss")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.title("Training vs Validation Loss")

plt.subplot(1,2,2)
plt.plot(train_r2, label="Train R²")
plt.plot(val_r2, label="Val R²")
plt.xlabel("Epochs")
plt.ylabel("R² Score")
plt.legend()
plt.title("Training vs Validation Accuracy (R²)")

plt.show()
```

**outline to predict custom value**

```python
def predict_custom(input_dict, model, label_encoders, scaler, num_cols, cat_cols, y_le, device):
    """
    Predicts the target value for a single custom input dictionary.
    
    Args:
        input_dict (dict): A dictionary containing the input features.
        model (nn.Module): The trained PyTorch model.
        label_encoders (dict): Dictionary of fitted LabelEncoders for features.
        scaler (StandardScaler): The fitted StandardScaler.
        num_cols (list): List of numerical column names.
        cat_cols (list): List of categorical column names.
        y_le (LabelEncoder): The fitted LabelEncoder for the target variable.
        device (torch.device): The device (CPU or GPU) to use for prediction.
        
    Returns:
        str: The predicted crop type.
    """
    # Create a DataFrame from the input dictionary
    df_new = pd.DataFrame([input_dict])

    # Ensure all original columns are present
    all_cols = num_cols + cat_cols
    missing_cols = set(all_cols) - set(df_new.columns)
    if missing_cols:
        raise ValueError(f"Missing columns in input_dict: {missing_cols}")

    df_new = df_new[all_cols]

    # Encode categorical features
    for col in cat_cols:
        if col in df_new.columns:
            le = label_encoders[col]
            # Handle new, unseen categories by assigning a dummy value (-1)
            try:
                df_new[col] = le.transform(df_new[col].astype(str))
            except ValueError:
                print(f"Warning: Category '{df_new[col].iloc[0]}' for column '{col}' not seen during training. Setting to -1.")
                df_new[col] = -1

    # Scale numerical features
    df_new[num_cols] = scaler.transform(df_new[num_cols])

    # Convert the preprocessed DataFrame to a PyTorch tensor
    X_new = torch.tensor(df_new.values, dtype=torch.float32).to(device)

    # Perform prediction
    model.eval()
    with torch.no_grad():
        preds_logits = model(X_new)
        # Get the index of the class with the highest logit
        pred_class_idx = preds_logits.argmax(dim=1).cpu().numpy()[0]
    
    # Decode the predicted class index back to the original string label
    predicted_crop = y_le.inverse_transform([pred_class_idx])[0]
    return predicted_crop

# Example of a custom input dictionary.
# Note that "Crop Type" is not included as it is the target variable to be predicted.
custom_input = {
    "Region": "West",
    "Soil Type": "Sandy",
    "Rainfall_mm": 800,
    "Temperature(c)": 27,
    "Fertilizer_Used": False,
    "Irrigation_Used": True,
    "Weather_Condition": "Cloudy",
    "Days_to_Harvest": 120,
    "Humidity": 65,
    "Moisture": 45,
    "Nitrogen": 20,
    "Potassium": 2,
    "Phosphorous": 15,
    "Fertilizer Name": "DAP",
    "Yield_tons_per_hectare" : 6.5
}

# The predict_custom function will now use all the input features to predict the 'Crop Type'
try:
    prediction = predict_custom(custom_input, model, label_encoders, scaler, num_cols, cat_cols, y_le, device)
    print(f"\nPredicted Crop Type: {prediction}")
except ValueError as e:
    print(f"Error: {e}")
```

# Why FT-Transformer 
🔹 Why FT-Transformer?

Your dataset is tabular with a mix of categorical (Region, Soil Type, Fertilizer Name) and numerical (Rainfall, Temperature, Nitrogen, etc.) features.
Traditional models like Random Forest, XGBoost, or Linear Regression work well, but:

They don’t fully capture interactions between categorical & numerical features.

They struggle when categorical features have many unique categories (e.g., crop types, fertilizer names).

The FT-Transformer (Feature Tokenizer Transformer) is designed specifically for tabular data, using the Transformer’s self-attention mechanism to:

Encode both categorical and numerical features into embeddings (like in NLP).

Learn complex feature interactions automatically.

Generalize well, especially when dataset size is moderate.

✅ Advantages of FT-Transformer

Handles Mixed Data (Cat + Num)

Works seamlessly with both categorical and continuous features.

Avoids manual heavy feature engineering.

Captures Complex Relationships

Self-attention allows the model to learn non-linear interactions between features (e.g., how rainfall interacts with soil type).

Better with High-Cardinality Categories

Instead of one-hot encoding (which explodes dimensionality), it uses embeddings for categorical features, just like word embeddings in NLP.

Scalable & Flexible

Can scale to large datasets and adapt to various tasks (classification, regression, multi-task).

Regularization via Transformer Architecture

Dropout + LayerNorm help avoid overfitting compared to deep MLPs on tabular data.

❌ Disadvantages of FT-Transformer

Data Hungry

Transformers generally need more data than XGBoost or Random Forest to reach peak performance.

With very small datasets (<2k rows), tree-based models often outperform.

Computationally Expensive

Training is slower than Random Forest or XGBoost.

Needs GPU for efficiency (especially with deeper architectures).

Hyperparameter Sensitive

Performance depends a lot on tuning (layers, heads, embedding size, dropout).

May require more experiments than simpler models.

Less Interpretable

Unlike decision trees, it’s harder to explain why the model made a prediction (though SHAP/LIME can help).

```python
def predict_custom(input_dict, model_path, scaler_path, le_path, y_le_path, num_features, num_classes, device):
    """
    Loads the trained model and preprocessing objects to predict the target value
    for a single custom input dictionary.
    
    Args:
        input_dict (dict): A dictionary containing the input features.
        model_path (str): The file path to the saved PyTorch model state.
        scaler_path (str): The file path to the saved StandardScaler.
        le_path (str): The file path to the saved feature LabelEncoders.
        y_le_path (str): The file path to the saved target LabelEncoder.
        num_features (int): The number of features in the input data.
        num_classes (int): The number of classes in the target variable.
        device (torch.device): The device (CPU or GPU) to use for prediction.
        
    Returns:
        str: The predicted crop type.
    """
    # Load preprocessing objects
    scaler = joblib.load(scaler_path)
    label_encoders = joblib.load(le_path)
    y_le = joblib.load(y_le_path)

    # Re-initialize the model with the correct architecture and load the saved state
    model = FTTransformer(num_features, num_classes).to(device)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()

    # Create a DataFrame from the input dictionary
    df_new = pd.DataFrame([input_dict])

    # Identify categorical and numerical columns from the loaded data
    cat_cols_loaded = list(label_encoders.keys())
    # Assuming numerical columns are the rest
    all_cols = num_cols + cat_cols_loaded
    missing_cols = set(all_cols) - set(df_new.columns)
    if missing_cols:
        raise ValueError(f"Missing columns in input_dict: {missing_cols}")

    df_new = df_new[all_cols]

    # Encode categorical features
    for col in cat_cols_loaded:
        if col in df_new.columns:
            le = label_encoders[col]
            try:
                df_new[col] = le.transform(df_new[col].astype(str))
            except ValueError:
                print(f"Warning: Category '{df_new[col].iloc[0]}' for column '{col}' not seen during training. Setting to -1.")
                df_new[col] = -1

    # Scale numerical features
    df_new[num_cols] = scaler.transform(df_new[num_cols])

    # Convert the preprocessed DataFrame to a PyTorch tensor
    X_new = torch.tensor(df_new.values, dtype=torch.float32).to(device)

    # Perform prediction
    with torch.no_grad():
        preds_logits = model(X_new)
        pred_class_idx = preds_logits.argmax(dim=1).cpu().numpy()[0]
    
    # Decode the predicted class index back to the original string label
    predicted_crop = y_le.inverse_transform([pred_class_idx])[0]
    return predicted_crop
```
